Для того чтобы заполнить данные равномерно, попробуем выявить зависимости для стобцов, представляющих особую важность. Для этого используем функцию, которая ищет строки с наибольшей корреляцией 

```python
def try_corr(my_column, my_df=df):
    temp_dict = {}
    for column in my_df.columns:
        try:
            temp_dict[column] = my_df[my_column].corr(my_df[column])
        except:
            pass 
    return pd.Series(temp_dict).sort_values(ascending=False)
```

Первым делом посмотрим на корреляции с высотой потолков:

```python
try_corr('ceiling_height')
```

Значения корреляции по модулю слишком малы. Придется заполнять пропуски медианами без дифференциации по другим признакам.

```python
def fill_by_median(column):
    print()
    print('Пропусков "до" в', column, ':', df[column].isna().sum())
    temp_median = df[column].median()
    df[column] = df[column].fillna(temp_median) 
    print('Пропусков "после" в', column, ':', df[column].isna().sum())
    print('_______________________')
```

<b>Жилая площадь и кухня</b>

На этом моменте возникает ряд небольших, но все же проблем. Если заполнять площадь кухни и жилую площадь медианами, то может получиться так, что суммарная площадь кухни и жилой площади будет больше общей площади. Попробуем избежать такого исхода.

```python
df.living_area.corr(df.rooms)
```

```python
def fill_by_rooms_median(column):
    print()
    print('Пропусков "до" в', column, ':', df[column].isna().sum())
    for rooms in df.rooms.unique():
        temp_median = df.loc[df['rooms'] == rooms, column].median()
        df.loc[df['rooms'] == rooms, column] = df.loc[df['rooms'] == rooms, column].fillna(temp_median) 
    print('Пропусков "после" в', column, ':', df[column].isna().sum())
    print('_______________________')
```

```python
fill_by_rooms_median('living_area')
fill_by_rooms_median('days_exposition')
```

```python
def fill_kitchen_area(row):
    return row['total_area'] - row['living_area']
```

```python
df.loc[df['kitchen_area'].isna(), 'kitchen_area'] = df[df['kitchen_area'].isna()].apply(fill_kitchen_area, axis=1)
```

### Цена квадратного метра

Такую цену посчитать очень просто: нужно разделить цену квардтиры на общую площадь. Но прежде проверим, что в значениях нужных нам переменных нет аномалий и выбросов 

```python
def price_per_m(row):
    return row['last_price'] / row['total_area']
```

```python
df['price_per_m'] = df.apply(price_per_m, axis=1)
df['price_per_m'].head()
```

### День недели, месяц и год публикации объявления

```python
df['exp_weekday'] = df['first_day_exposition'].dt.weekday
df['exp_month'] = df['first_day_exposition'].dt.month
df['exp_year'] = df['first_day_exposition'].dt.year
```

### Этаж квартиры; варианты — первый, последний, другой

```python
def floor_cat(row):
    tem_total_floors = row['floors_total']
    temp_floor = row['floor']
    if temp_floor == 1:
        return 'Первый'
    elif temp_floor == tem_total_floors:
        return 'Последний'
    else:
        return 'Другой'
```

```python
df['floor_category'] = df.apply(floor_cat, axis=1)
```

```python
df['floor_category'].value_counts()
```

### Соотношение жилой и общей площади, а также отношение площади кухни к общей

```python
def share_of_living(row):
    return row['living_area'] / row['total_area']

def share_of_kitchen(row):
    return row['kitchen_area'] / row['total_area']
```

```python
df['living_share'] = df.apply(share_of_living, axis=1)
df['kitchen_share'] = df.apply(share_of_kitchen, axis=1)
```

```python
df.head()
```

Теперь проверим те самые квартиры, где кухня и жилая площадь в сумме дают значения больше, чем 1.

```python
df.query('living_share + kitchen_share > 1')
```

Таких строк всего 73, их можно безболезненно отбросить

```python
print(df.shape[0])
df = df.query('~(living_share + kitchen_share > 1)')
print(df.shape[0])
```

```python
def show_data(column, my_bins=100, my_df=df):
    my_df.boxplot(column, vert=False, figsize=(20, 2))
    plt.title('Распределение величины ' + str(column), fontdict={'fontsize': 20})
    plt.xlabel('value')
    plt.show()
    my_df.hist(column, figsize=(20, 5), bins=my_bins)
    plt.xlabel(column)
    plt.ylabel('count')
    plt.title('Распределение величины ' + str(column), fontdict={'fontsize': 20})
    plt.show()
    
    
def show_good_data(column, my_bins=100, my_df=df):
    q1 = my_df[column].quantile(0.25)                 
    q3 = my_df[column].quantile(0.75)
    iqr = q3 - q1
    temp_min = q1 - 1.5 * iqr
    temp_max = q3 + 1.5 * iqr
    df_temp = my_df[(my_df[column] <= temp_max) & (my_df[column] >= temp_min)]
    df_temp.boxplot(column, vert=False, figsize=(20, 2))
    plt.title('Распределение величины ' + str(column), fontdict={'fontsize': 20})
    plt.xlabel('value')
    plt.show()
    plt.show()
    df_temp.hist(column, figsize=(20, 5), bins=my_bins)
    plt.xlabel(column)
    plt.ylabel('count')
    plt.title('Распределение величины ' + str(column), fontdict={'fontsize': 20})
    plt.show()
    print('Доля хороших данных', df_temp.shape[0] / df.shape[0])
    return (temp_min, temp_max)
    
```
