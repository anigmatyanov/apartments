```python
def try_corr(my_column, my_df=df):
    temp_dict = {}
    for column in my_df.columns:
        try:
            temp_dict[column] = my_df[my_column].corr(my_df[column])
        except:
            pass 
    return pd.Series(temp_dict).sort_values(ascending=False)
```

```python
def floor_cat(row):
    tem_total_floors = row['floors_total']
    temp_floor = row['floor']
    if temp_floor == 1:
        return '–ü–µ—Ä–≤—ã–π'
    elif temp_floor == tem_total_floors:
        return '–ü–æ—Å–ª–µ–¥–Ω–∏–π'
    else:
        return '–î—Ä—É–≥–æ–π'
```

–ü–µ—Ä–≤—ã–º –¥–µ–ª–æ–º —Å–æ–∑–¥–∞–¥–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å —Ü–µ–Ω–æ–π –≤ –º–∏–ª–ª–∏–æ–Ω–∞—Ö, —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ –≤–∏–¥–∞

‚úçÔ∏è –î–∞–ª–µ–µ –Ω–∞–ø–∏—à–∏–º —Ñ—É–Ω–∫—Ü–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–º–æ–≥—É—Ç –∏–∑—É—á–∞—Ç—å –¥–∞–Ω–Ω—ã–µ:

```python
def show_data(column, my_bins=100, my_df=df):
    my_df.boxplot(column, vert=False, figsize=(20, 2))
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ–ª–∏—á–∏–Ω—ã ' + str(column), fontdict={'fontsize': 20})
    plt.xlabel('value')
    plt.show()
    my_df.hist(column, figsize=(20, 5), bins=my_bins)
    plt.xlabel(column)
    plt.ylabel('count')
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ–ª–∏—á–∏–Ω—ã ' + str(column), fontdict={'fontsize': 20})
    plt.show()
    
    
def show_good_data(column, my_bins=100, my_df=df):
    q1 = my_df[column].quantile(0.25)                 
    q3 = my_df[column].quantile(0.75)
    iqr = q3 - q1
    temp_min = q1 - 1.5 * iqr
    temp_max = q3 + 1.5 * iqr
    df_temp = my_df[(my_df[column] <= temp_max) & (my_df[column] >= temp_min)]
    df_temp.boxplot(column, vert=False, figsize=(20, 2))
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ–ª–∏—á–∏–Ω—ã ' + str(column), fontdict={'fontsize': 20})
    plt.xlabel('value')
    plt.show()
    plt.show()
    df_temp.hist(column, figsize=(20, 5), bins=my_bins)
    plt.xlabel(column)
    plt.ylabel('count')
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ–ª–∏—á–∏–Ω—ã ' + str(column), fontdict={'fontsize': 20})
    plt.show()
    print('–î–æ–ª—è —Ö–æ—Ä–æ—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö', df_temp.shape[0] / df.shape[0])
    return (temp_min, temp_max)
    
```

3. –ú–æ–∂–Ω–æ –ø–æ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å, –∫–æ–≥–¥–∞ –≥—Ä–∞—Ñ–∏–∫ —Å–∏–ª—å–Ω–æ –º–µ–Ω—è–µ—Ç—Å—è –∏ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–∞–Ω–æ–≤—è—Ç—Å—è –ø–æ—Ö–æ–∂–∏–º–∏ –Ω–∞ "–Ω–µ—Ç–∏–ø–∏–ø—á–Ω—ã–µ"
